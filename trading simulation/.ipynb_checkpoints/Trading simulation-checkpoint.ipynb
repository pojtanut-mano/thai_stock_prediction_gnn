### Stock price data

```python
def get_stock_prices():
    
    extended_path = r"..\data\raw\price"
    folder_path = os.path.join(extended_path)
    arr = os.listdir(folder_path)
    
    prices = {}
    for index, name in enumerate(arr):
        file_path = os.path.join(extended_path, name)
        prices[name[:-4]] = pd.read_csv(file_path)
    
    return prices
```

### Classification

```python
def get_predict_class(folder_name, start_period, num_period, input_type, output_type):
    
    extended_path = r"data"
    folder_path = os.path.join(extended_path, folder_name)
    arr = os.listdir(folder_path)
    
    if input_type in ['train', 'test', 'valid']:
        if output_type == 'class':
            file_name = input_type + r'_pred_df.csv'
        elif output_type == 'prob':
            file_name = input_type + r'_pred_proba_df.csv'
        elif output_type == 'label':
            file_name = input_type + r'_label_df.csv'
        else:
            raise ValueError("Wrong input: 'output_type' must be 'class', 'prob', or 'label'")
    else:
        raise ValueError("Wrong input: 'input_type' must be 'train', 'test', or 'valid'")
    
    df = pd.DataFrame()
    output_path = r'raw_output/' + file_name
    for i in range(start_period ,start_period+num_period):
        
        sub_folder_path = r'period_'+str(i)
        file_path = os.path.join(folder_path, sub_folder_path, output_path)
        
        temp = pd.read_csv(file_path)
        if i == start_period:
            df = temp
        else:
            df = pd.concat([df, temp], ignore_index=True)
   
    return df
```

### Regression


```python
def get_predict_reg(folder_name, start_period, num_period, input_type, output_type):
    
    extended_path = r"data"
    folder_path = os.path.join(extended_path, folder_name)
    arr = os.listdir(folder_path)
    
    if input_type in ['train', 'test', 'valid']:
        if output_type == 'reg':
            file_name = input_type + r'_pred_df.csv'
        elif output_type == 'label':
            file_name = input_type + r'_label_df.csv'
        else:
            raise ValueError("Wrong input: 'output_type' must be 'reg', or 'label'")
    else:
        raise ValueError("Wrong input: 'input_type' must be 'train', 'test', or 'valid'")
    
    df = pd.DataFrame()
    output_path = r'raw_output/' + file_name
    for i in range(start_period ,start_period+num_period):
        
        sub_folder_path = r'period_'+str(i)
        file_path = os.path.join(folder_path, sub_folder_path, output_path)
        
        temp = pd.read_csv(file_path)
        if i == start_period:
            df = temp
        else:
            df = pd.concat([df, temp], ignore_index=True)
   
    return df
```

# Trading simulation function

```python
def plot_invest(history_invest, name):
    day = [i for i in range(len(history_invest))]
    plt.plot(day,history_invest)
    plt.title(name)
    plt.xlabel('day')
    plt.ylabel('value')
    plt.show()
```

```python
def check_ticker(ticker, market, date, prices):
    
    if (ticker in market) and (ticker in prices): 
        stock = prices[ticker]
        if stock[stock['Date']==date].shape[0] == 1:
            return True
    return False
```

```python
def get_trade_name(class_value, class_prob, reg_value, reg_label, market, stock_number, selection_method, date, prices):

    trade_name = []
    
    if selection_method == 'class':
        new_reg_value = dict([(y,x) for x,y in reg_value])
        for val, ticker in class_prob:
            if (class_value[ticker] == '2') and check_ticker(ticker, market, date, prices):
                if reg_value != []:
                    trade_name.append((ticker, 0))
                else:
                    trade_name.append((ticker, val))
            if len(trade_name) == stock_number:
                break
        return trade_name
    
    if selection_method == 'reg':
        for val, ticker in reg_value:
            if (val>0) and check_ticker(ticker, market, date, prices):
                trade_name.append((ticker,val))
            if len(trade_name) == stock_number:
                break
        return trade_name
    
    if selection_method == 'class+reg':
        for val, ticker in reg_value:
            if (class_value[ticker] == '2') and (val > 0) and check_ticker(ticker, market, date, prices): 
                trade_name.append((ticker,val))
            if len(trade_name) == stock_number:
                break
        return trade_name
    
    if selection_method == 'reg_label':
        for val, ticker in reg_label:
            if (val>0) and check_ticker(ticker, market, date, prices):
                trade_name.append((ticker,val))
            if len(trade_name) == stock_number:
                break
        return trade_name
```

```python
def get_expected_profit(stock, pricing_method, MA_day, val, idx):
    
    if pricing_method == 'reg':
        return 1 + float(val)
    
    if pricing_method == 'MA':
        MA = 0
        for i in range(1, 1 + MA_day):
            MA += stock.loc[idx-i,'High']/stock.loc[idx-i,'Open']
        MA = max(0,(MA/MA_day-1))+1
        return MA
```

```python
def trading_simulation(prices, initial_fund, class_values, class_probs, class_names, reg_values, 
                       reg_labels, market, stock_number, selection_method, pricing_method, MA_day,
                      use_class, use_reg, investing_method, cut_loss_threshold, verbose, priority, oc_price):

    history_invest = [initial_fund]
    all_traded_name = {}
    
    for idx in range(len(class_values)):

        if use_class == True:
            class_value = dict(np.stack((class_names[1:], class_values[idx][1:]), axis = -1))
            class_prob = np.stack((class_probs[idx][1:], class_names[1:]), axis = -1).tolist()
            class_prob = [(float(x), y) for x,y in class_prob]
            class_prob.sort(reverse = True)
        else:
            class_value = []
            class_prob = []
            
        if use_reg == True:
            reg_value = np.stack((reg_values[idx][1:], class_names[1:]), axis = -1).tolist()
            reg_label = np.stack((reg_labels[idx][1:], class_names[1:]), axis = -1).tolist()
            reg_value = [(float(x), y) for x,y in reg_value]
            reg_label = [(float(x), y) for x,y in reg_label]
            reg_value.sort(reverse = True)
            reg_label.sort(reverse = True)
        else:
            reg_value = []
            reg_label = []
        
        date = class_values[idx][0]
        trade_name = get_trade_name(class_value, class_prob, reg_value, reg_label, market, 
                                    stock_number, selection_method, date, prices)
        
        if verbose>=1: 
            print(date)
        
        total_invest = 0
        for ticker, val in trade_name:
        
            if ticker not in all_traded_name:
                all_traded_name[ticker] = 0
            all_traded_name[ticker] += 1

            stock = prices[ticker]
            if investing_method == 'fix':
                invest = initial_fund/len(trade_name)
            elif investing_method == 'acc':
                invest = history_invest[-1]/len(trade_name)
            else:
                raise ValueError("Wrong input: 'investing_method' must be 'fix', or 'acc'")
                
            idx = stock[stock['Date']==date].index[0]
            expected_profit = get_expected_profit(stock, pricing_method, MA_day, val, idx)
            
            if oc_price == 'open':
                open_price = float(stock[stock['Date']==date]['Open'])
            elif oc_price == 'close':
                open_price = float(stock.loc[idx-1,'Close'])
            else:
                raise ValueError("Wrong input: 'oc_price' must be 'open', or 'close'")
            
            high_price = float(stock[stock['Date']==date]['High'])
            low_price = float(stock[stock['Date']==date]['Low'])
            close_price = float(stock[stock['Date']==date]['Close'])

            unit = invest/(open_price)
            sell_limit = open_price*expected_profit
            cut_loss = open_price*cut_loss_threshold

            if verbose>=2:
                print('ticker = {0:8}: open price = {1:6.2f}, expected profit = {2:6.4f}\n \
                        high = {3:6.4f}, low = {4:6.4f}, close = {5:6.4f}'.format(
                        ticker, open_price, expected_profit, high_price/open_price, 
                        low_price/open_price, close_price/open_price))
            
            if priority == 'low':
                if low_price < cut_loss:
                    if verbose>=3: print('\tcut_loss: profit = {0:.2f}'.format(cut_loss*unit-invest))
                    invest = cut_loss*unit
                elif high_price > sell_limit:
                    if verbose>=3: print('\tsell: profit = {0:.2f}'.format(sell_limit*unit-invest))
                    invest = sell_limit*unit    
                else:
                    if verbose>=3: print('\tclose: profit = {0:.2f}'.format(close_price*unit-invest))
                    invest = (close_price)*unit
            elif priority == 'high':
                if high_price > sell_limit:
                    if verbose>=3: print('\tsell: profit = {0:.2f}'.format(sell_limit*unit-invest))
                    invest = sell_limit*unit    
                elif low_price < cut_loss:
                    if verbose>=3: print('\tcut_loss: profit = {0:.2f}'.format(cut_loss*unit-invest))
                    invest = cut_loss*unit
                else:
                    if verbose>=3: print('\tclose: profit = {0:.2f}'.format(close_price*unit-invest))
                    invest = (close_price)*unit
            else:
                raise ValueError("Wrong input: 'priority' must be 'high', or 'low'")

            total_invest += invest
        
        if len(trade_name) == 0:
            history_invest.append(history_invest[-1])
        else:
            if investing_method == 'fix':
                history_invest.append(history_invest[-1] + total_invest - initial_fund)
            elif investing_method == 'acc':
                history_invest.append(total_invest)
        
        if verbose>=1: print('................', history_invest[-1], '.................')
    
    return history_invest, all_traded_name
```

```python
folder_class_name = '3_6_2021_new'
folder_reg_name = ''
start_period = 11
num_period = 5
input_type = 'test'

if folder_class_name != '':
    predict_class = get_predict_class(folder_class_name, start_period, num_period, input_type, 'class')
    predict_class_prob = get_predict_class(folder_class_name, start_period, num_period, input_type, 'prob')
    predict_class_label = get_predict_class(folder_class_name, start_period, num_period, input_type, 'label')
    class_values = predict_class.values.tolist()
    class_probs = predict_class_prob.values.tolist()
    class_probs = [[re.match(r'^\[[ ]*[0-9\.\-e\+]+[ ]*[0-9\.\-e\+]+[ ]*([0-9\.\-e\+]+)[ ]*\]$',x).group(1) if len(x)>12 else x for x in y] for y in class_probs]
    class_names = list(predict_class.columns)
    class_labels = predict_class_label.values.tolist()
else:
    class_values = []
    class_probs = []
    class_names = []

if folder_reg_name != '':
    predict_reg = get_predict_reg(folder_reg_name, start_period, num_period, input_type, 'reg')
    predict_reg_label = get_predict_reg(folder_reg_name, start_period, num_period, input_type, 'label')
    reg_values = predict_reg.values.tolist()
    reg_labels = predict_reg_label.values.tolist()
    reg_names = list(predict_class.columns)
else:
    reg_values = []
    reg_labels = []
    reg_names = []

if class_names == []:
    class_names = reg_names
```

```python
initial_fund = 10000
market = SET # SET, SET100, all_stock
stock_number = 5
selection_method = 'class' # 'class', 'reg', 'class+reg', 'reg_label'
pricing_method = 'MA' # 'MA', 'reg'
MA_day = 5
use_class = True
use_reg = False
investing_method = "acc" # 'fix', 'acc'
cut_loss_threshold = 0.7
verbose = 3 # 0, 1, 2, 3
priority = 'low' # 'high', 'low'
oc_price = 'close' # 'open', 'close'

history_invest,all_traded_name = trading_simulation(prices, initial_fund, class_values, class_probs, class_names, reg_values, 
                        reg_labels, market, stock_number, selection_method, pricing_method, MA_day,
                        use_class, use_reg, investing_method, cut_loss_threshold, verbose, priority, oc_price)

plot_invest(history_invest, selection_method+'/'+pricing_method+'/'+investing_method)
print(Counter(all_traded_name))
```
